### submitting first list in the name of list1 to the program

```python
import matplotlib as mp
 
string=input('Enter your string')

list1=string.split(' ')
print(len(list1))


def sortbyorder(m):
  n = len(m)
  strng=[]
  for e in m:                                                                            
    l = list(e)
    is_sorted = False
    while not is_sorted:
        for i in range(len(l)-1):
            if l[i] > l[i+1]:
                l[i+1], l[i] = l[i], l[i+1]
                break
        else:
            is_sorted=True
    strng.append("".join(l))
  print(strng)
  return strng


def method():
    sortedstring=list1
    s_sort = sortbyorder(sortedstring)
    xaxis = []
    yaxis = []
    n = 0
    for i in range(len(s_sort)):
        xaxis.append(n)
        n = n + 1
    for s in s_sort:
        yaxis.append(len(s))
    import matplotlib.pyplot as plt
    plt.scatter(xaxis, yaxis)
    plt.xlabel("Order")
    plt.ylabel("length")
    plt.show()
     
        
method()
```
