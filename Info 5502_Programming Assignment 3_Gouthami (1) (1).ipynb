### Importing all required Libraries

```python
import numpy
import matplotlib.pyplot as plt
import array
```

### Write a Python program to simulate the calculation of π using Monte Carlo method

```python
list=10**8
x=numpy.random.uniform(low=-1,high=1,size=[list,1])
y=numpy.random.uniform(low=-1,high=1,size=[list,1])
value=x**2+y**2<1
aprroximate_value_of_pi=4*numpy.sum(value)/list
print('pi: {}, pi approximate value: {:.3f}'.format(numpy.pi,aprroximate_value_of_pi))
```

### Your program draws the circle and the square, and plot the random points in the graph as shown in Figure 1. You need simulate different numbers of points (at least include 10, 500, 1,000, 5,000, 10,000, 50,000, 80,000, 100,000, 200,000) and calculate the value π for each different number of points.

```python
random_points =[5,10, 500, 1000, 5000, 10000, 20000, 50000, 80000, 100000, 200000]
a = []
for Np in random_points:
  x = numpy.random.uniform(low=0, high=1, size=[Np, 1])
  y = numpy.random.uniform(low=0, high=1, size=[Np, 1])
  border = (x-0.5)**2+(y-0.5)**2<0.25
  pi_value = 4*numpy.sum(border)/Np
  a.append(4*numpy.sum(border)/Np)
  x_in = x[border]
  y_in = y[border]
  create_fig=plt.figure(figsize=(5,5))
  s_plot = create_fig.add_subplot(1, 1, 1)
  box=plt.Rectangle((0,0), 1, 1,edgecolor='red',fill=False)
  border=plt.Circle((0.5,0.5),radius=0.5,edgecolor='purple',fill=False)
  s_plot.add_patch(box)
  s_plot.add_patch(border)
  plt.scatter(x, y, s=5)
  plt.scatter(x_in, y_in, color='r', s=5)
  plt.title(r"$n = {}, \pi approximate value = {:.5f}$".format(Np,  pi_value))
plt.show()
```

### Then draw a line chart to show the change of the accuracy of the value of π calculated when the number of points is increased.

```python

k = numpy.array(random_points)
l = numpy.array(a)
plt.xlabel('Number of points')
plt.ylabel('pi approximate value')
plt.title('Pi accuracy change')
plt.plot(k,l,color = 'Green')
plt.show()
```
